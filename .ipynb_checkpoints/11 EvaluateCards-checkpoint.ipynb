###### 使用随机森林进行缺失值处理


```python
%%time
from sklearn.ensemble import RandomForestRegressor as RFR

def fill_missing_rf(x,y,to_fill):
    """
    x:需要填充的特征矩阵
    y:完整的，没有缺失值的标签
    to_fill:需要填充的列
    """
    # 构建新的特征矩阵
    df=x.copy()
    fill=df.loc[:,to_fill]
    df=pd.concat([df.loc[:,df.columns!=to_fill],pd.DataFrame(y)],axis=1)
    
    # 找出训练集和测试集
    y_train=fill[fill.notnull()]
    y_test=fill[fill.isnull()]
    x_train=df.iloc[y_train.index,:]
    x_test=df.iloc[y_test.index,:]
    
    # 用随机森林填补缺失值
    rfr=RFR(n_estimators=100)
    rfr=rfr.fit(x_train,y_train)
    y_predict=rfr.predict(x_test)
    return y_predict

x=data.iloc[:,1:]
to_fill='MonthlyIncome'
y=data['SeriousDlqin2yrs'] #y=data.iloc[:,0]

y_pred=fill_missing_rf(x,y,to_fill)       
```

```python
# 用于判读每个分箱中是否存在正负样例
# 如果有一个不存在就向后合并
def sure_01(num_bins):
    for i in range(len(num_bins)):
        if 0 in num_bins[0][2:]:
            num_bins[0:2]=[(
                num_bins[0][0]
                ,num_bins[1][1]
                ,num_bins[0][2]+num_bins[1][2]
                ,num_bins[0][3]+num_bins[1][3]
            )]
            continue
        for i in range(len(num_bins)):
            if 0 in num_bins[i][2:]:
                num_bins[i-1:i+1]=[(
                    num_bins[i-1][0]
                    ,num_bins[i][1]
                    ,num_bins[i-1][2]+num_bins[i][2]
                    ,num_bins[i-1][3]+num_bins[i][3]
                )]
                break
        else:
            break
        # 如果原样本中第一行与第二行都没有包含正样例，或者没有包含负样例，合并之后也不会包含
        # 因此需要在检查，第一组是否已经包含了两种样例
    
```

##### IV=sum((good%-bad%)*WOE(i))
##### WOE(i)=ln(good%/bad%)
##### WOE越大表示该分箱中的优质客户越多，IV用于对整个特征进行描述
###### IV<0.03 特征几乎不带有效信息
######  IV~0.03-0.09 有效信息少
###### IV~0.1-0.29 有效信息一般
######  IV~0.3-0.49 有效信息较多
###### IV>=0.5 有效信息非常多，存在可疑

```python
# 定义WOE和IV函数
def WOE(num_bins):
    columns=['min','max','count_0','count_1']
    df=pd.DataFrame(num_bins,columns=columns)
    
    df['total']=df.count_0+df.count_1
    df['percentage']=df.total/df.total.sum() #一个样本所占比例
    df['bad_rate']=df.count_1/df.total
    df['good%']=df.count_0/df.count_0.sum()
    df['bad%']=df.count_1/df.count_1.sum()
    df['woe']=np.log(df['good%']/df['bad%'])
    return df

df_temp=WOE(num_bins)

# 计算IV
def IV(data):
    rate=data['good%']-data['bad%']
    iv=np.sum(rate*data['woe'])
    return iv

iv=IV(df_temp)
iv
```

```python
# 利用K方检验，合并箱体，画出箱子数目对应的IV值的曲线图
import matplotlib.pyplot as plt
import scipy

# 指定最小的箱子数目n
def plot_bins_iv(num_bins,n):
    iv=[]
    axisx=[]
    num_bins_=num_bins.copy()
    while len(num_bins_)>n:
        pvs=[]
        # 获取num_bins_两两之间的卡方检验置信度（或卡方检验）
        for i in range(len(num_bins_)-1):
            x1=num_bins_[i][2:]
            x2=num_bins_[i+1][2:]
            # 参数为0返回chi2值，为1返回p值
            pv=scipy.stats.chi2_contingency([x1,x2])[1]
            # ci2=scipy.stats.chi2_contingncy([x1,x2])[0]
            pvs.append(pv)
        # 通过p值进行处理，合并p值大的两组
        i=pvs.index(max(pvs))
        #print("pvs:",max(pvs),i)
        num_bins_[i:i+2]=[(
            num_bins_[i][0],
            num_bins_[i+1][1],
            num_bins_[i][2]+num_bins_[i+1][2],
            num_bins_[i][3]+num_bins_[i+1][3]
        )]
        
        bins_df=WOE(num_bins_)
        axisx.append(len(num_bins_))
        iv.append(IV(bins_df))
        
    plt.figure()
    plt.plot(axisx,iv)
    plt.xticks(axisx)
    plt.xlabel('length of num_bins_')
    plt.ylabel('IV of the num_bins_')
    plt.show()
    
plot_bins_iv(num_bins,2)
```

###### 从上图中选择下降最快的转折点，当分箱数目为6时，效果最佳

```python
# 输入参数为：原始箱数，想要保留的箱子数目
def get_bins(num_bins,n):
    num_bins_=num_bins.copy()
    while len(num_bins_)>n:
        pvs=[]
        # 获取num_bins_两两之间的卡方检验置信度（或卡方检验）
        for i in range(len(num_bins_)-1):
            x1=num_bins_[i][2:]
            x2=num_bins_[i+1][2:]
            # 参数为0返回chi2值，为1返回p值
            pv=scipy.stats.chi2_contingency([x1,x2])[1]
            # ci2=scipy.stats.chi2_contingncy([x1,x2])[0]
            pvs.append(pv)
        # 通过p值进行处理，合并p值大的两组
        i=pvs.index(max(pvs))
        #print("pvs:",max(pvs),i)
        num_bins_[i:i+2]=[(
            num_bins_[i][0],
            num_bins_[i+1][1],
            num_bins_[i][2]+num_bins_[i+1][2],
            num_bins_[i][3]+num_bins_[i+1][3]
        )]
    return num_bins_

# 获取合并之后的箱子
after_bins=get_bins(num_bins,6)
after_bins
```

```python
# 计算woe值
bins_df=WOE(after_bins)
bins_df
```

###### 将选取最佳分箱个数的函数打包为一个函数

```python
def graphforbestbins(DF,x,y,n=5,q=20,graph=True):
    """
    自动最优分箱函数，基于卡方检验的分箱
    参数：
    DF：原始数据集
    x：需要分箱的别名
    y：分箱个数对应的标签y列名
    n：保留的分箱个数
    q：初始分箱数目
    graph：是否绘制IV图像
    """
    # 获得原始分箱
    DF=DF[[x,y]].copy()
    DF['qcut'],bins=pd.qcut(DF[x],retbins=True,q=q,duplicates="drop")
    count_y0=DF.loc[DF[y]==0].groupby(by="qcut").count()[y]
    count_y1=DF.loc[DF[y]==1].groupby(by="qcut").count()[y]
    num_bins=[*zip(bins,bins[1:],count_y0,count_y1)]
    # 将只存在正样例或负样例的样本向后进行合并
    for i in range(q):
        if 0 in num_bins[2:]:
            num_bins[0:2]=[(
                num_bins[i][0],
                num_bins[i+1][1],
                num_bins[i][2]+num_bins[i+1][2],
                num_bins[i][3]+num_bins[i+1][3]
            )]
        for i in range(len(num_bins)):
            if 0 in num_bins[i][2:]:
                num_bins[i-1:i+1]=[(
                    num_bins[i-1][0],
                    num_bins[i][1],
                    num_bins[i-1][2]+num_bins[i][2],
                    num_bins[i-1][3]+num_bins[i][3]
                )]
                break
        else:
            break
        
    after_bins=get_bins(num_bins,n)
    # 计算分箱后的woe值
    bins_df=WOE(after_bins)
    # 计算iv值
    iv=IV(bins_df)
    
    # 是否绘制IV图像
    if graph:
        plot_bins_iv(num_bins,n)
    return bins_df  

model_data.columns
# graphforbestbins(model_data,x,y,n=5,q=20,graph=True)            
```

```python
bins_of_col={}
y=model_data.columns[0]
# 生成自动分箱的分箱区间和分箱后的iv值
for col in auto_col_bins:
    bins_df=graphforbestbins(model_data
                            ,col
                            ,y
                            ,n=auto_col_bins[col] 
                            ,q=20 #原始分箱数目
                            ,graph=False 
                           )
    bins_list=sorted(set(bins_df["min"]).union(bins_df["max"]))
    # 保证区间覆盖使用np.inf和-np.inf替换最大和最小值
    bins_list[0],bins_list[-1]=-np.inf,np.inf
    bins_of_col[col]=bins_list
    
# 合并手动分箱数据
bins_of_col.update(hand_bins)
bins_of_col.keys()
```

### 制作评分卡

##### 公式score=A-B*log(odds)
###### A，B为常数，A为补偿，B为刻度，log(odds)表示一个人违约的可能性
###### A，B两个常数满足假设：某个特定违约概率下的预期分值；指定的违约概率翻倍的分数（PDO）
###### 600=A-B*log(1/60)
###### 620=A-B*log(1/30)

```python
B=(20)/np.log(2)
A=600+np.log(1/60)*B
B,A
```

```python
# 基础分：不受评分卡中各个特征影响的基础分，就是将截距log(odds)带入公式计算
# 截距为：lr.intercept_
base_score=A-B*lr.intercept_
base_score

model_data.head()
```
