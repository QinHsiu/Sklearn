```python
# 实例：泰坦尼克幸存者的预测
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

data=pd.read_csv(r'data/titanitic/mytrain.csv')
# 删除无关多余的列
data=data.loc[ : , ~data.columns.str.contains("^Unnamed")]
data.info()
```

```python
data.head(5)
```

```python
data.drop(['Cabin','Name','Ticket'],inplace=True,axis=1)
#等价与
# data=data.drop(['Cabin'],inplace=False,axis=1)
```

```python
# 处理缺失值
data['Age']=data['Age'].fillna(data['Age'].mean())
data.info()
```

```python
# 删除有缺失值的行
data=data.dropna()
data.info()
```

```python
labels=data['Embarked'].unique().tolist()
data['Embarked']=data['Embarked'].apply(lambda x:labels.index(x))
data
```

```python
# data['Sex']=(data['Sex']=="male").astype("int") #不使用loc和iloc
data.loc[:,'Sex']=(data['Sex']=="male").astype("int") #使用列名
data.iloc[:,3] #使用列索引
```

```python
data.head()
```

```python
# 获取特征和目标值
x=data.iloc[:,data.columns!="Survived"]
y=data.iloc[:,data.columns=="Survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
# print(x_train.shape,y_train.shape)

x_train.index=range(x_train.shape[0])

# 将索引按顺序排列
for i in [x_train,x_test,y_train,y_test]:
    i.index=range(i.shape[0])
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
def classify(x_train,y_train,x_test,y_test):
    clf=DecisionTreeClassifier(random_state=25)
    clf.fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    return score

classify(x_train,y_train,x_test,y_test)
```

```python
# 使用交叉验证，取均值
from sklearn.model_selection import cross_val_score
clf=DecisionTreeClassifier(random_state=25)
score=cross_val_score(clf,x,y,cv=10).mean()
score
```

```python
tr=[]
te=[]
for i in range(10):
    clf=DecisionTreeClassifier(random_state=25
                               ,max_depth=i+1
                               ,criterion="entropy"
                              )
    clf=clf.fit(x_train,y_train)
    score_tr=clf.score(x_train,y_train)
    score_te=cross_val_score(clf,x,y,cv=10).mean()
    tr.append(score_tr)
    te.append(score_te)

plt.plot(range(1,11),tr,color="blue",label="train")
plt.plot(range(1,11),te,color="red",label="test")
plt.legend()
plt.show()
```

```python
import numpy as np
# 生成长度为50的，取值为0~0.5的有序列表
gini_threholds=np.linspace(0,0.5,20)
# print(gini_threholds)

# 网格搜索
from sklearn.model_selection import GridSearchCV
parameters={
    "criterion":("gini","entropy")
    ,"splitter":("best","random")
    ,"max_depth":[*range(1,10)]
    ,"min_samples_leaf":[*range(1,50,5)]
    ,"min_impurity_decrease":[*gini_threholds]
    
}


# 网格搜素：能够同时增加多个参数的技术，枚举技术
def gridSearch():
    clf=DecisionTreeClassifier(random_state=25)
    GS=GridSearchCV(clf,parameters
                    ,cv=10
                    #,scoring='neg_mean_squared_error'
                   )
    GS=GS.fit(x_train,y_train)
    score=GS.score(x_test,y_test)
    # 返回最佳参数组合
    # GS.best_params_
    # 返回最佳组合下的score
    # GS.best_score_
    # 返回最好的模型
    # grid_search.best_estimator_
    return score

gridSearch()
 
```
