```python
# 单层线性回归神经网络
def LinearR(x1,x2):
    w1,w2,b=0.15,0.15,-0.2
    z=x1*w1+x2*w2+b
    return z

LinearR(X[:,0],X[:,1])
```

```python
# 单层分类神经网络
# 阈值
y_and=[0,0,0,1]

def sigmoid(z):
    return 1/(1+np.exp(-z))

def and_sigmoid(x1,x2):
    w1,w2,b=0.15,0.15,-0.2
    z=x1*w1+x2*w2+b
    # 回归转化为分类
    o=sigmoid(z)
    y_sigmoid=[int(x) for x in o>=0.5]
    y_sign=[int(x) for x in z>=0]
    return o,y_sigmoid,y_sign

o,y_sigm,y_sign=and_sigmoid(X[:,0],X[:,1])
y_sigm
y_sign
```

```python
# 分支方法：信息熵、基尼指数
# criterion:entropy,gini
from sklearn import tree
from sklearn.datasets import load_wine
wine=load_wine()
wine.data
wine.target
```

```python
import pandas as pd
# 拼接数据集
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)

wine.feature_names
wine.target_names
```

```python
from sklearn.model_selection import train_test_split
# 划分数据集为测试集与验证集,使用留余法,随机划分
x_train,x_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.3)
x_train.shape
```

```python
y_train.shape
```

```python
clf=tree.DecisionTreeClassifier(criterion="entropy") #实例化
clf.fit(x_train,y_train)# 训练模型
clf.score(x_test,y_test)# 导入测试集，进行评价
```

```python
import graphviz
feature_name=wine.feature_names
class_name=wine.target_names
#print(feature_name,class_name)
dot_data=tree.export_graphviz(clf,
                             feature_names=feature_name,
                             class_names=class_name,
                             filled=True,#填充颜色
                             rounded=True)# 圆角

graph=graphviz.Source(dot_data)
graph
```

```python
clf.feature_importances_
{*zip(feature_name,clf.feature_importances_)}
```

```python
# random_state用来设置分枝中的随机模式的参数，默认为None，设置值后，会一直长出同一棵树，
# 让模型稳定下来

# splitter用来控制决策树的随机选项，两种值：best和random
clf=tree.DecisionTreeClassifier(criterion="entropy",
                                random_state=30,
                               splitter="random") #实例化
clf.fit(x_train,y_train)# 训练模型
clf.score(x_test,y_test)# 导入测试集，进行评价
clf.score(x_train,y_train)# 导入训练集，进行评价
```

```python
# 剪枝参数
# 设置最大树高 max_depth
# 设置满足分支条件的最小训练样本数：min_samples_leaf,如果节点分支后样本数小于该值就不分枝
# 设置节点至少包含的样本数：min_samples_split
# splitter用来控制决策树的随机选项，两种值：best和random
# max_features限制分枝时考虑的特征个数
# min_impurity_decrease #限制信息增益的大小

clf=tree.DecisionTreeClassifier(criterion="entropy",
                                random_state=30,
                               splitter="random",
                               max_depth=3,
                                min_samples_split=10,
                                min_samples_leaf=10
                               ) #实例化
clf.fit(x_train,y_train)# 训练模型
clf.score(x_test,y_test)# 导入测试集，进行评价
clf.score(x_train,y_train)# 导入训练集，进行评价


import graphviz
feature_name=wine.feature_names
class_name=wine.target_names
#print(feature_name,class_name)
dot_data=tree.export_graphviz(clf,
                             feature_names=feature_name,
                             class_names=class_name,
                             filled=True,#填充颜色
                             rounded=True)# 圆角

graph=graphviz.Source(dot_data)
graph
```

```python
# max_features限制分枝时考虑的特征个数
# min_impurity_decrease #限制信息增益的大小

import matplotlib.pyplot as plt

test=[]
for i in range(10):
    clf=tree.DecisionTreeClassifier(criterion="entropy",
                                random_state=30,
                               splitter="random",
                               max_depth=i+1
                                #min_samples_split=10,
                                #min_samples_leaf=10
                               ) #实例化
    clf=clf.fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    test.append(score)
    
plt.plot(range(1,11),test,color="red",label="max_depth")
plt.legend()
plt.show()
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
import pandas as pd
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons,make_circles,make_classification
from sklearn.tree import DecisionTreeClassifier

X,y=make_classification(n_samples=100,
                       n_features=2,
                       n_redundant=0, # 添加冗余特征
                        n_informative=2, #包含信息的特征
                        random_state=1, #随机模式
                        n_clusters_per_class=1 #每个簇内包含的标签数
                       )
# 查看生成数据集
X.shape
y.shape
plt.scatter(X[:,0],X[:,1])
```

```python
rng=np.random.RandomState(2) #生成一种随机模式
X += 2*rng.uniform(size=X.shape) #加减0~1之间的随机数
linearly_separable=(X,y) #生成新的X

plt.scatter(X[:,0],X[:,1])
```

```python
# 使用make_moons创建月亮型数据 ，使用make_circles创建环形数据，
# 将三组数据打包放在datasets中
datasets=[make_moons(noise=0.3,random_state=0),make_circles(noise=0.2,factor=0.5,random_state=1),
         linearly_separable]
datasets
```

```python
# 创建画布
figure=plt.figure(figsize=(6,9))
i=1

for ds_index,ds in enumerate(datasets):
    X,y=ds
    X=StandardScaler().fit_transform(X)
    x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.4,
                                                  random_state=42)
    # 找出数据集中两个特征的最大值和最小值，最大值加0.5,最小值减去0.5
    x1_min,x1_max=X[:,0].min()-0.5,X[:,0].max()+0.5
    x2_min,x2_max=X[:,1].min()-0.5,X[:,1].max()+0.5
    
    # 该函数用于生成网格数据，能够将两个一维数组生成两个二维矩阵
    array1,array2=np.meshgrid(np.arange(x1_min,x1_max,0.2),
                              np.arange(x2_min,x2_max,0.2))
    
    
           
    # 生成彩色画布
    cm=plt.cm.RdBu
    cm_bright=ListedColormap(["#FF0000","#0000FF"])

    ax=plt.subplot(len(datasets),2,i)

    if ds_index==0:
        ax.set_title("Input data")

    # 训练集    
    ax.scatter(x_train[:,0],x_train[:,1],c=np.squeeze(y_train),cmap=cm_bright,edgecolors="k")
    # 测试机
    ax.scatter(x_test[:,0],x_test[:,1],c=np.squeeze(y_test),cmap=cm_bright,alpha=0.6,edgecolors="k")

    ax.set_xlim(array1.min(),array1.max())
    ax.set_ylim(array2.min(),array2.max())
    ax.set_xticks(())
    ax.set_yticks(())

    i+=1
    
    ax=plt.subplot(len(datasets),2,i)
    clf=DecisionTreeClassifier(max_depth=5)
    clf.fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    
    z=clf.predict_proba(np.c_[array1.ravel(),array2.ravel()])[:,1]
    
    # np.c_[np.array([1,2,3,4]),np.array([2,3,4,5])]
    z=z.reshape(array1.shape)
    ax.contourf(array1,array2,z,cmap=cm,alpha=0.8)
    
    # 训练集    
    ax.scatter(x_train[:,0],x_train[:,1],c=y_train,cmap=cm_bright,edgecolors="k")
    # 测试机
    ax.scatter(x_test[:,0],x_test[:,1],c=y_test,cmap=cm_bright,alpha=0.6,edgecolors="k")
    
    ax.set_xlim(array1.min(),array1.max())
    ax.set_ylim(array2.min(),array2.max())
    
    # 设置坐标不显示标尺和数字
    ax.set_xticks(())
    ax.set_yticks(())
    if ds_index==0:
        ax.set_title('Decision Tree')
        
    ax.text(array1.max()-0.3,array2.min()+0.3,
            ('{:.1f}%'.format(score*100)),size=15,
           horizontalalignment="right")
    i+=1
    
plt.tight_layout()
plt.show()



```
