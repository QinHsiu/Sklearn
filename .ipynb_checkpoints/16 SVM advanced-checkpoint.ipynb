```python
# 使用支持向量机进行二分类

# 导入模块
from sklearn.svm import SVC
from sklearn.datasets import make_circles,make_moons,make_blobs,make_classification
from sklearn import svm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.colors import ListedColormap #带色彩的画布
```

```python
# 创建数据集，定义核函数的选择

n_samples=100 #创建样本数

datasets=[
    make_moons(n_samples=n_samples,noise=0.2,random_state=0)
    ,make_circles(n_samples=n_samples,noise=0.2,factor=0.5,random_state=1)
    ,make_blobs(n_samples=n_samples,centers=2,random_state=2)
    ,make_classification(n_samples=n_samples,n_features=2,n_informative=2,n_redundant=0,random_state=3)
    ]

# 核函数
kernel=['linear','poly','sigmoid','rbf']




# 创建子图
nrows=len(datasets)
ncols=len(kernel)+1
fig,axes=plt.subplots(nrows,ncols,figsize=(20,10))

for ds_cnt,(x,y) in enumerate(datasets):
    # 在图像的第一列放置原始数据的分布
    ax=axes[ds_cnt,0]
    if ds_cnt==0:
        ax.set_title("Input data")
    ax.scatter(x[:,0],x[:,1],c=y,zorder=10,cmap=plt.cm.Paired,edgecolors="k")
    ax.set_xticks([])
    ax.set_yticks([])
    
    # 第二层循环：在不同的核函数中循环
    # 从图像第二列开始
    for est_idx,k in enumerate(kernel):
        # 定义图的位置
        ax=axes[ds_cnt,est_idx+1]
        
        # 建模
        clf=SVC(kernel=k,gamma=2).fit(x,y)
        score=clf.score(x,y)
        
        # 绘制图像本身分布散点图
        ax.scatter(x[:,0],x[:,1],c=y,zorder=10,cmap=plt.cm.Paired,edgecolor="k")
        
        # 绘制支持向量
        ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=50,facecolor='none',edgecolors="k")
        

        
        # 获取最大，最小值
        x_min,x_max=x[:,0].min()-.5,x[:,0].max()+.5
        y_min,y_max=x[:,1].min()-.5,x[:,1].max()+.5
        # 绘制决策边界
        # 找网格
        # 一次性使用最大值和最小值生成网格
        # 表示为[起始值：阶数值：步长]
        # 如果步长为复数，则其整数部分就是起始值和结束值之间创建数据的数量（点的个数）
        xx,yy=np.mgrid[x_min:x_max:200j,y_min:y_max:200j]
        # np.c_类似于np,vstack()
        z=clf.decision_function(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    
        # 填充等高线不同区域的颜色
        # 为z>0绘制一种颜色
        ax.pcolormesh(xx,yy,z>0,cmap=plt.cm.Paired)
        
        #绘制等高线
        ax.contour(xx,yy,z,colors=['k','k','k'],linestyles=['--','-','--'],levels=[-1,0,1])
        
        
        
        # 设置标题以及不显示坐标
        ax.set_title("%s:%.2f"%(k,score))
        ax.set_xticks([])
        ax.set_yticks([])

        # 设置标题,第一行
        #if ds_cnt==0:
        #    ax.set_title(k)
            
        # 为每张图添加分类的分数
        ax.text(0.95,0.06,('%.2f'%score).lstrip('0')
               ,size=15
                ,bbox=dict(boxstyle="round",alpha=0.8,facecolor="white")# 为分数添加一个白色格子用于显示
                ,transform=ax.transAxes #确定文字所对应的坐标轴，就是ax子图的坐标轴本身
                ,horizontalalignment='right' #位于坐标轴什么方向
               )
        
plt.tight_layout()
plt.show()

        
```

##### 二分类SVC种的样本不均衡问题：重要参数为class_weight

##### 在样本不均衡的时候可以采用下采样和上采样的方法填充和删除数据，但这样会增加SVC的计算时间
##### 可以使用SVC类中的class_weight和接口fit种可以设定的sample_weight
##### 在逻辑回归中可以使用字典形式为每个特征设置权重，也可以通过设置balanced模式，直接使用
##### n_samples/(n_classes*np.bincount(y))作为权重
##### 但在SVM中，分类判断是基于决策边界的，而最终决定用怎样的支持向量机和决策边界的参数是参数C，
##### 所以所有样本均衡是通过C来进行调整的

### SVC的参数

##### 1.class_weight
##### 可以输入字典形式、balanced或者不填，默认为None
##### 对于SVC，将类i参数设置为class_weight[i]*C，如果没有给出具体的class_weight，则所有类都被假定相同的权重1
##### 模型会根据数据原本的状况进行训练，可以通过输入字典形式{"标签值":权重}的形式进行调整
##### 使用balanced模式，这个模式使用y的值进行自动调整输入数据中类的频率：n_samples/(n_classes*np.bincount(y))

##### 2.sample_weight
##### 数组形式，结构为(n_samples),必须对应输入fit中的特征的每个样本
##### 每个样本在fit时的权重，让权重*每个样本对应的C值来迫使分类器强调设定的权重更大的样本
##### 如果同时设置了两个参数，则C会同时受到两个参数的影响，class_weight中设定的权重*sample_weight中设定的权重*C

```python
# 导入所需库
from sklearn.svm import SVC
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
class_1=500 #类1有500个样本
class_2=50 #类2有50个样本

centers=[[0.0,0.0],[2.0,2.0]] #设置两个类别的中心
clusters_std=[1.5,0.5] #设定两个类别的方差，通常来说量比较大的会更加松散

x,y=make_blobs(n_samples=[class_1,class_2]
               ,centers=centers
              ,cluster_std=clusters_std
               ,random_state=0
               ,shuffle=False
              )


plt.scatter(x[:,0],x[:,1],c=y,cmap="rainbow",s=50)
plt.show()
```

```python
# 绘制两个模型下的决策边界
plt.figure(figsize=(6,5))
plt.scatter(x[:,0],x[:,1],c=y,cmap="rainbow",s=10)
ax=plt.gca()

# 设置x取值边界
xlim=ax.get_xlim()
ylim=ax.get_ylim()

xx=np.linspace(xlim[0],xlim[1],30)
yy=np.linspace(ylim[0],ylim[1],30)
YY,XX=np.meshgrid(yy,xx)
xy=np.vstack([XX.ravel(),YY.ravel()]).T

# 未设置class_weight
clf=SVC(kernel="linear",C=1.0)
clf.fit(x,y)
# 设置class_weight
class_weight={1:10}
zlf=SVC(kernel="linear",class_weight=class_weight,C=1.0)
zlf.fit(x,y)


# 第二部找出样本点到决策边界的距离
z_clf=clf.decision_function(xy).reshape(XX.shape)
a=ax.contour(XX,YY,z_clf,colors="black",levels=[0],alpha=0.5,linestyles=['-'])

z_wlf=zlf.decision_function(xy).reshape(XX.shape)
b=ax.contour(XX,YY,z_wlf,colors="red",levels=[0],alpha=0.5,linestyles=['-'])

# 第三步绘制图例
plt.legend([a.collections[0],b.collections[0]],["non class_weight","class_weight"],
          loc="upper right")#两条线

plt.show()
```

##### 使用新的模型评估指标进行评价

##### 混淆矩阵，confuseMatrix
##### 基于混淆矩阵有6个评价指标
##### 正确率：所有判断正确的样本数/总的样本数
##### 精确度，查准率：P=TP/(TP+FP)
##### 召回率：R=TP/(TP+FN)
##### F-score=2*P*R/(P+R)=2/(1/p+1/R)，偏向于数值更小的一方
##### 特异度(specificity)：表示样本中所有真实为0的样本中被正确预测为0的所占比例,s=TN/(TN+FN)
##### 假正率：1-特异度，衡量一个模型将多数类别判错的能力，FN/(TN+FN)

##### 使用不同阈值进行预测,找到一个最好的recall和precision

```python
# 自建数据集
class_1=7
class_2=4
centers_=[[0,0],[1,1]]
clusters_std=[0.5,1]
x_,y_=make_blobs(n_samples=[class_1,class_2],centers=centers_,cluster_std=clusters_std
                 ,random_state=0,shuffle=False)





# 使用逻辑回归展示阈值
from sklearn.linear_model import LogisticRegression as LR

clf_to=LR().fit(x_,y_)
prob=clf_to.predict_proba(x_) #得出预测的概率

# 将样本和概率放到一个DataFrame中
import pandas as pd
prob=pd.DataFrame(prob)
prob.columns=[0,1]



def PR(threshold,x_,y_):
    p=[]
    r=[]
    for t in threshold:
        # 使用阈值进行判断
        for i in range(prob.shape[0]):
            if prob.loc[i,1]>t:
                prob.loc[i,'pred']=1
            else:
                prob.loc[i,'pred']=0
        # 将正确标签添加到矩阵中
        prob['y_true']=y_
        p_0=P(prob.loc[:,'y_true'],prob.loc[:,'pred'],labels=[1,0]) #必须先填真实值
        r_0=R(prob.loc[:,'y_true'],prob.loc[:,'pred'],labels=[1,0]) #必须先填真实值
        p.append(p_0)
        r.append(r_0)
    #print(p,r)
    plt.plot(threshold,r,c="r",label="recall")
    plt.plot(threshold,p,c="b",label="precision")
    fpr=list(map(lambda x:1-x,p))
    plt.plot(fpr,r,c="green",label="x_fpr and y_tpr")
    plt.xlabel('threshold')
    plt.ylabel('recall and precision')
    plt.legend()
    plt.show()
    
threshold=np.linspace(0.25,0.6,10)
PR(threshold,x_,y_)
```

```python
# 自制数据集
class_1=500
class_2=50
centers=[[0,0],[2,2]]
clusters_std=[1.5,0.5]

x,y=make_blobs(n_samples=[class_1,class_2],centers=centers,cluster_std=clusters_std
              ,random_state=1,shuffle=False)


plt.scatter(x[:,0],x[:,1],c=y,s=30,cmap="rainbow")
plt.show()

clf_prob=SVC(kernel="linear",C=1.0,probability=True).fit(x,y)
clf_prob.predict_proba(x) #生成各类标签下的概率
```

### SVN也可以生成概率，可以通过设置阈值来进行判断
##### Platt缩放中涉及交叉验证对于大型数据集代价大，计算缓慢

```python
# FPR=FP/(FP+TN)
cm[1,0]/cm[1,:].sum()
```

```python
# TPR，recall=TP/(TP+FN)
cm[0,0]/cm[0,:].sum()
```

```python
# 开始绘图
def PR(threshold,x,y):
    prob=SVC(kernel="linear",random_state=0,probability=True).fit(x,y)
    prob=prob.predict_proba(x)
    prob=pd.DataFrame(prob)
    prob.columns=[0,1]
    
    prob['y_true']=y
    
    p=[]
    r=[]
    
    recall=[]
    f=[] ##fpr
    for t in threshold:
        # 使用阈值进行判断
        for i in range(prob.shape[0]):
            if prob.loc[i,1]>t:
                prob.loc[i,'pred']=1
            else:
                prob.loc[i,'pred']=0
        # 将正确标签添加到矩阵中
        p_0=P(prob.loc[:,'y_true'],prob.loc[:,'pred'],labels=[1,0]) #必须先填真实值
        r_0=R(prob.loc[:,'y_true'],prob.loc[:,'pred'],labels=[1,0]) #必须先填真实值
        p.append(p_0)
        r.append(r_0) #使用自带函数求解
        
        # 使用混淆矩阵
        cm=CM(prob.loc[:,'y_true'],prob.loc[:,'pred'],labels=[1,0])
        recall.append(cm[0,0]/cm[0,:].sum())
        f.append(cm[1,0]/cm[1,:].sum())
        
        
    # plt.plot(threshold,r,c="r",label="recall/tpr")
    # plt.plot(threshold,p,c="b",label="precision")
    fpr=list(map(lambda x:1-x,p))
    fpr.sort()
    r.sort()
    
    #recall.sort()
    #f.sort()
    
    # plt.plot(threshold,fpr,c="gray",label="fpr")
    plt.plot(f,recall,c="green",label="roc")
    plt.plot(fpr,r,c="r",label="x_fpr and y_tpr")
    plt.plot(thresholds+0.05,thresholds+0.05,c="b",linestyle="--")
    plt.xlabel('fpr')
    plt.ylabel('recall/tpr')
    plt.legend()
    plt.show()
    
thresholds=np.linspace(clf_prob.predict_proba(x)[:,1].min(),clf_prob.predict_proba(x)[:,1].max(),50,endpoint=False)
PR(thresholds,x,y)
```
