### 预测明天是否会下雨


```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
path=os.getcwd()+"\\data\\weather\\"
filename=os.listdir(path)
filename

for f in filename:
    if f[-4:]=='.csv':
        data=pd.read_csv(path+f)
        break
data.info() #查看数据信息
```

```python
%%time
# 根据当天降雨量对当天降雨与否进行缺失值填充
x_train.loc[x_train['Rainfall']>=1,'RainToday']='Yes'
x_train.loc[x_train['Rainfall']<1,'RainToday']='No'
x_train.loc[x_train['Rainfall']==np.nan,'RainToday']=np.nan

x_test.loc[x_test['Rainfall']>=1,'RainToday']='Yes'
x_test.loc[x_test['Rainfall']<1,'RainToday']='No'
x_test.loc[x_test['Rainfall']==np.nan,'RainToday']=np.nan
```

### 使用支持向量机进行建模

```python
from time import time
import datetime
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score,recall_score
from sklearn.model_selection import cross_val_score

# 需要对测试集进行降维
y_test=y_test.iloc[:,0].ravel()
y_train=y_train.iloc[:,0].ravel()
```

```python
# 对比各种核函数的跑分（观察数据是否线性可分）
for kernel in ['linear','poly','rbf','sigmoid']:
    begin=time()
    clf=SVC(kernel=kernel,gamma="auto"
           ,degree=1 #多项式最高次数
            ,cache_size=5000
           ).fit(x_train,y_train)
    result=clf.predict(x_test)
    score=clf.score(x_test,y_test) #返回accuracy
    recall=recall_score(y_test,result) #第一个参数为真实值，第二个参数为预测值
    auc=roc_auc_score(y_test,clf.decision_function(x_test)) #使用距离作为阈值
    print("%s testing accuracy is %f ,recall is %f,auc is %f"%(kernel,score,recall,auc))
    print("time cost: ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
    
```

### 调参数的策略：
##### 1.追求recall
##### 2.追求最高预测准确率
##### 3.达到两者之间的平衡

```python
# 使用随机森林进行预测
from sklearn.ensemble import RandomForestClassifier as RFC

for estimators in range(50,201,50):
    begin=time()
    rf=RFC(n_estimators=estimators,random_state=420).fit(x_train,y_train)
    score=rf.score(x_test,y_test)
    result=rf.predict(x_test)
    recall=recall_score(y_test,result)
    #auc=roc_auc_score(y_test,result) 
    print(" n_estimaters :%d ,accuracy is %.4f ,recall is %.4f"%(estimators,score,recall))
    print("time cost ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
```

### 对支持向量机模型进行调参
### 1.追求recall

```python
# 加入参数class_weight="balanced"
# class_weight 取{1:10}表示类别1:10，隐藏了类别0:1这个比例（更加重视少数类）

for kernel in ['linear']:
    begin=time()
    clf=SVC(kernel=kernel,gamma="auto"
           ,degree=1 #多项式最高次数
            ,class_weight="balanced"
            ,cache_size=5000
           ).fit(x_train,y_train)
    result=clf.predict(x_test)
    score=clf.score(x_test,y_test) #返回accuracy
    recall=recall_score(y_test,result) #第一个参数为真实值，第二个参数为预测值
    auc=roc_auc_score(y_test,clf.decision_function(x_test)) #使用距离作为阈值
    print("%s testing accuracy is %f ,recall is %f,auc is %f"%(kernel,score,recall,auc))
    print("time cost: ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
    
```

### 2.追求准确率

```python
# 2.追求准确率
valuec=pd.Series(y_test).value_counts()
valuec
```

```python
# 查看模型的特异值
from sklearn.metrics import confusion_matrix as CM

# 线性核函数进行预测
clf=SVC(kernel="linear",gamma="auto"
       ,cache_size=5000).fit(x_train,y_train)

result=clf.predict(x_test)
cm=CM(y_test,result,labels=[0,1]) #混淆矩阵
cm
```

```python
# 求特异值
specificity=cm[1,1]/cm[1,:].sum()
specificity #几乎所有的0都被判断正确了，还有不少1也判断正确
```

```python
# 可以通过调class_weight参数来进行比较
for i in np.linspace(0.01,0.05,10):
    begin=time()
    clf=SVC(kernel="linear"
           ,gamma="auto"
            ,cache_size=5000
            ,class_weight={1:1+i}
           ).fit(x_train,y_train)
    result=clf.predict(x_test)
    score=clf.score(x_test,y_test)
    recall=recall_score(y_test,result)
    auc=roc_auc_score(y_test,clf.decision_function(x_test))
    print("class_weight 1:%.4f , score is %.4f, recall is %.4f, auc is %.4f"%(i+1,score,recall,auc))
    print("time cost ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
    
    
```

```python
# 3.追求平衡
# 调整核函数的c值来追求平衡
import matplotlib.pyplot as plt

c_range=np.linspace(0.01,20,30) #惩罚项

recallall=[]
aucall=[]
scoreall=[]

for c in c_range:
    begin=time()
    clf=SVC(kernel="linear",C=c,cache_size=5000
           ,class_weight="balanced"
           ).fit(x_train,y_train)
    result=clf.predict(x_test)
    score=clf.score(x_test,y_test)
    recall=recall_score(y_test,result)
    auc=roc_auc_score(y_test,clf.decision_function(x_test))
    aucall.append(auc)
    recallall.append(recall)
    scoreall.append(score)
    
plt.plot(c_range,scoreall,c="red",label="accuracy")
plt.plot(c_range,recallall,c="b",label="recall")
plt.plot(c_range,auc,c="gray",label="auc")
plt.legend()
plt.show()
```

##### 通过绘制ROC曲线来选择好的阈值

```python
%%time
# 使用之前所调的最好的参数的模型

clf=SVC(kernel="linear",C=3.1663
        ,cache_size=5000
        ,class_weight="balanced"
       ).fit(x_train,y_train)

from sklearn.metrics import roc_curve as ROC
import matplotlib.pyplot as plt



FPR,Recall,thresholds=ROC(y_test,clf.decision_function(x_test),pos_label=1)
area=roc_auc_score(y_test,clf.decision_function(x_test))


plt.figure() #创建画布
plt.plot(FPR,Recall,color="red"
         ,label="roc curve(area=%.2f)"%area
        )
plt.plot([0,1],[0,1],color="black",linestyle="--")
plt.xlim([-0.05,1.05])
plt.ylim([-0.05,1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("Receiver operating characterstic example")
plt.legend(loc="lower right")
plt.show()
```

```python
# 根据所得recall和fpr找到最佳阈值
maxindex=(Recall-FPR).tolist().index(max(Recall-FPR))
# 最佳阈值
thresholds[maxindex]
```

```python
# 基于最佳阈值，来确定y_predict,并确定该阈值下的recall和FPR
from sklearn.metrics import accuracy_score as AC

begin=time()
clf=SVC(kernel="linear",C=3.1663
       ,cache_size=5000
       ,class_weight="balanced").fit(x_train,y_train)

# 手动根据阈值进行划分
prob=pd.DataFrame(clf.decision_function(x_test))
prob.loc[prob.loc[:,0]>=thresholds[maxindex],'y_pred']=1
prob.loc[prob.loc[:,0]<thresholds[maxindex],'y_pred']=0

prob.loc[:,"y_pred"].isnull().sum()
```

```python
begin=time()
# 检查本身的准确性
score=AC(y_test,prob.loc[:,"y_pred"].values)
recall=recall_score(y_test,prob.loc[:,"y_pred"])
print("testing accuracy is %f,recall is %f"%(score,recall))
print("time cost is ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
```
