```python
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor

boston=load_boston()
regressor=DecisionTreeRegressor(random_state=0)
# 10折交叉验证
cross_val_score(regressor,boston.data,boston.target,cv=10
                ,scoring="neg_mean_squared_error" #评价标准
               )
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng=np.random.RandomState(1) #设置随机数种子（随机数生成范围为[0,1]）

x=np.sort(5*rng.rand(80,1),axis=0)

y=np.sin(x).ravel() #降维

plt.figure()
plt.scatter(x,y,s=20,edgecolor="blue",c="darkorange",label="data")
plt.show()

# np.random.random((2,1))
# np.sin(x).ravel().shape #降维
```

```python
# 加入噪声
y[::5]=3*(0.5-rng.rand(16))

plt.figure()
plt.scatter(x,y,s=20,edgecolor="blue",c="darkorange",label="data")
plt.show()
```

```python
max_depth=[2,5]

x_test=np.arange(0,5,0.01)[:,np.newaxis] #升维

color=['red','green']
# print(x_test.shape,x_test.ravel().shape)

def plot(x,y,x_test,y_p,i):
    plt.figure()
    plt.scatter(x,y,s=20,edgecolor="blue",c="darkorange",label="data")
    plt.plot(x_test,y_p,color=color[i],label="max_depth_ %d"%max_depth[i],linewidth=2)
    plt.xlabel('data')
    plt.ylabel("target")
    plt.title("Decision Tree Regression")
    plt.legend()
    plt.show()
    
for i in range(len(max_depth)):
    regr=DecisionTreeRegressor(max_depth=max_depth[i])
    regr.fit(x,y)
    y_p=regr.predict(x_test)
    plot(x,y,x_test,y_p,i)
    

# 出现过拟合现象
    
```
