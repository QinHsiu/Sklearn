```python
# 导入库和数据集
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd


from time import time
import datetime
import matplotlib.pyplot as plt
```

```python
data=load_breast_cancer()
x=data.data #特征
y=data.target #标签
x.shape
```

```python
np.unique(y) #查看标签取值（二分类）
```

```python
plt.scatter(x[:,0],x[:,1],c=y) #前两个特征所画的图
plt.show()                
```

```python
# 使用PCA降维
from sklearn.decomposition import PCA 
x_dr=PCA(2).fit_transform(x)

plt.scatter(x_dr[:,0],x_dr[:,1],c=y) #降至二维后的图像
plt.show()
```

```python
# 划分训练集和测试集
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# 使用四个核函数查看得分
kernel=['linear','poly','sigmoid','rbf']

for k in kernel:
    begin=time()
    clf=SVC(kernel=k
           ,gamma="auto"
            , degree=1 #默认非线性
            ,cache_size=5000 #使用最大内存 
           ).fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    print("The accuracy under kernel %s is %.2f"%(k,score))
    print("cost time: ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
```

```python
# 划分训练集和测试集
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)


# 使用四个核函数查看得分
kernel=['linear','poly','sigmoid','rbf']

for k in kernel:
    begin=time()
    clf=SVC(kernel=k
           ,gamma="auto"
            ,degree=1 #默认3,多项式的次数，选择'poly'核函数时才有效
            ,cache_size=5000 #使用最大内存 
           ).fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    print("The accuracy under kernel %s after standard is %.4f"%(k,score))
    print("cost time: ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
```

##### 对于多项式核函数,在高次项时计算非常缓慢
##### rbf核和多项式核都不擅长处理量纲不统一的数据集

##### 对rbf进行调参
##### gamma:默认值为auto,仅仅在非线性核函数情况下有效
##### auto:gamma=1/n_features
##### scale:gamma=1/(n_features*X.std())
##### auto_deprecated:表示没有传递明确的gamma值
##### coef0 浮点数，默认为0.0，核函数中的常数项，只在核函数为sigmoid和poly情况下有效

```python
scores=[]
gamma_range=np.logspace(-10,1,50) #返回在数值刻度上均匀间隔的数字


gammas=['auto','scale']

for i in gamma_range:
    begin=time()
    clf=SVC(kernel='rbf'
           ,gamma=i
            ,degree=1 #默认3,多项式的次数，选择'poly'核函数时才有效
            ,cache_size=5000 #使用最大内存 
           ).fit(x_train,y_train)
    score=clf.score(x_test,y_test)
    scores.append(score)
    #print("The accuracy under gamma %.2f after standard is %.4f"%(i,score))
    #print("cost time: ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
    
print("max score:%.3f,gamma is :%.3f"%(max(scores),gamma_range[scores.index(max(scores))]))
plt.plot(gamma_range,scores,c="r",linestyle="-")
plt.show()
```

### 使用网格搜索调多项式核函数的参数

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV #带交叉验证的网格搜索

begin=time()

gamma_range=np.logspace(-10,1,50)
coef0_range=np.linspace(0,5,10)

param_grid=dict(
    gamma=gamma_range
    ,coef0=coef0_range)

cv=StratifiedShuffleSplit(n_splits=5,test_size=3,random_state=420) #使用交叉验证，划分成5份
grid=GridSearchCV(SVC(kernel='poly',degree=1,cache_size=5000),param_grid=param_grid)
grid.fit(x,y)


print("The best parameters are %s with a score of %.5f"%(grid.best_params_,grid.best_score_))
print("cost time: ",datetime.datetime.fromtimestamp(time()-begin).strftime("%M:%S:%f"))
```

##### 两组数据完全线性可分，可以找出一个决策边界使得训练集上的分类误差为0，这种数据就称为存在”硬间隔“
##### 两组数据几乎时完全线性可分的，但决策边界在训练集上存在较小的训练误差，这种数据就称为存在”软间隔“

##### 引入松弛系数，对于重要参数C有
##### C为浮点数，默认为1，其值不小于0，表示对松弛系数的惩罚项，其值越大，表示越重视错分类项，会导致SVC选择边际小的

```python
# 调线性核函数
scores=[]

c_range=np.linspace(0.01,50,100)
for i in c_range:
    clf=SVC(kernel="rbf",C=i,cache_size=5000).fit(x_train,y_train)
    scores.append(clf.score(x_test,y_test))
    
print("max score is %.4f and C is %.3f"%(max(scores),c_range[scores.index(max(scores))]))
plt.plot(c_range,scores)
plt.show()
```
