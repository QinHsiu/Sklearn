### KMeans用于非结构化数据的矢量量化（VQ），降维（压缩样本的所含的信息量）
### 使用计算的质心来替代原有的数据

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin #对两个序列中的点进行距离匹配的函数
from sklearn.datasets import load_sample_image 
# 导入图片数据所需库
from sklearn.utils import shuffle #打乱
from time import time
```

```python
china=load_sample_image("china.jpg")
china.shape
# (长度、宽度、像素)
```

```python
# 先将图片里面的数据归一化
n_clusters=64

china=np.array(china,dtype=np.float64)/china.max() #数据归一化

w,h,d=original_shape=tuple(china.shape) # 原始数据的三维

assert d==3,"一个格子中的特征数目不等于3" #使用断言，如果不满足抛出异常,特征数目必须不变

image_array=np.reshape(china,(w*h,d)) #将数据降至二维
# np.reshape(需要改变的数据，需要形成的shape
image_array.shape
```

```python
# 矢量建模

# 先使用1000个数据找出质心
image_array_sample=shuffle(image_array,random_state=0)[:1000]
kmeans=KMeans(n_clusters=n_clusters,random_state=0).fit(image_array_sample)
kmeans.cluster_centers_.shape
```

```python
labels=kmeans.predict(image_array)
labels.shape #27W个样本所对应的簇的质心的索引
# set(labels)
```

```python
# 使用质心来替换所有的样本
image_kmeans=image_array.copy()
image_kmeans.shape

for i in range(w*h):
    image_kmeans[i]=kmeans.cluster_centers_[labels[i]]

# 查看生成的新图片的信息
pd.DataFrame(image_kmeans).drop_duplicates().shape
```

```python
# 恢复图像
image_kmeans=image_kmeans.reshape(w,h,d)
image_kmeans.shape
```

```python
plt.imshow(image_kmeans)
```

### 随机进行矢量量化

```python
centroid_random=shuffle(image_array,random_state=0)[0:n_clusters]
centroid_random.shape
```

```python
labels_random=pairwise_distances_argmin(centroid_random,image_array,axis=0)
# 函数pairwise_distances_argmin(x1,x2,axis)
# 用来计算x2中每个样本到x1中的每个样本的距离，并返回和x2相同形状的，与x1中对应的最近的样本的索引
labels_random.shape
```

```python
# 使用随机质心来替换所有样本
image_random=image_array.copy()

for i in range(w*h):
    image_random[i]=centroid_random[labels_random[i]]

# 还原图像
image_random=image_random.reshape(w,h,d)
plt.imshow(image_random)
```

```python
# 观察三种图的不同
plt.figure(figsize=(10,10))
plt.axis("off")
plt.title('original image')
plt.imshow(china)

plt.figure(figsize=(10,10))
plt.axis("off")
plt.title('quantized image(64 colors,Kmeans)')
plt.imshow(image_kmeans)


plt.figure(figsize=(10,10))
plt.axis("off")
plt.title('quantized image(64 colors,random)')
plt.imshow(image_random)

plt.show()
```
