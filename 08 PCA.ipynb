### 降维方法：PCA（主成分分析）和SVD（奇异值分解）
##### 特征选择后的特征具有可解释性，降维之后会生成新的特征（属于特征创造，具有不可解释性）
##### 线性回归不适合使用PCA

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
data=load_iris()
y=data.target
x=data.data
x.shape # 一个二维数组（a,b）
```

```python
import pandas as pd
pd.DataFrame(x) #四维特征矩阵
```

```python
# 使用降维，绘制散点图
pca=PCA(n_components=2) #降到2维
pca=pca.fit(x) #拟合模型
x_dr=pca.transform(x) #获取新矩阵
x_dr.shape #将特征降为二维特征

# 将数据集降维为二维特征矩阵
x_dr[y==0,0],x_dr[y==0,1]
def plt_iris(x):
    colors=['r','b','g']
    markers=['o','x','p']
    labels=['0','1','2']
    for i in range(len(colors)):
        plt.scatter(x[y==i,0]
                    ,x[y==i,1]
                    ,color=colors[i]
                    ,marker=markers[i]
                    ,label=labels[i]
                   ,alpha=.5 #设置透明度
                   )
    plt.legend(loc='lower right')
    plt.title('pca_2_dimensions')
    plt.xlabel('x_pca_0')
    plt.ylabel('x_pca_1')
    plt.show()
plt_iris(x_dr)
    
    
```

### 查看降维后每个新特征的信息量大小(可解释方差大小)


```python
pca.explained_variance_
```

### 查看降维后每个新特征向量所占的信息量占原始数据总信息量的百分比


```python
pca.explained_variance_ratio_
```

### 大部分特征集中在了降维后的第一个新特征上

```python
pca.explained_variance_ratio_.sum()
```

### 查看选择特征个数对信息占比的影响


```python
PCA().fit(x).explained_variance_ratio_
```

### 使用累加函数观察


```python
import numpy as np
np.cumsum(PCA().fit(x).explained_variance_ratio_)
```

### 绘制关于选择特征的学习曲线

```python
def p_l():
    plt.plot([1,2,3,4],np.cumsum(PCA().fit(x).explained_variance_ratio_))
    plt.xticks([1,2,3,4])
    plt.xlabel('number of components after dimension reduction')
    plt.ylabel('cumulative explained variance ')
    plt.show()
p_l()
```

### 从上图可以看出降到2或3维就是很好的降维选择

### 按信息量占比选超参数

##### n_components输入[0,1]之间的浮点数，并且让参数svd_solver=='full',表示希望降维后的总解释性方差占比最少阈值
##### 指定的百分比表示想要保留的信息量（想要保留的信息量大小）


```python
pca_f=PCA(n_components=0.97,svd_solver="full")
pca_f=pca_f.fit(x)
x_f=pca_f.transform(x)
x_f.shape
pca_f.explained_variance_ratio_.sum()
```

### 上述svd_solver取值及其含义如下
##### 默认值'auto'，基于shape和n_xomponents的默认策略来选择分解器，如果输入数据的特征数小于维度min(shape)的80%，就启用效率更高的“randomized”

##### 'full'，从scipy.linalg.svd中调用标准的LAPACK分解器来生成精确完整的SVD ,适合数据特征空间充足的情况

##### 'arpack'，从scipy.linalg.svds中调用ARPACK分解器来进行截断奇异值分解(svd trucated)分解时将特征数量降到n_components中输入数值k，可以加快运算角度，适合特征矩阵很大的时候， 一般用于特征矩阵为稀疏矩阵的情况

##### 'randomized',通过Halko等人的随机方法进行随机SVD。在'full'中，分解器会根据原始数据和输入的n_components值去计算和寻找符合需求的新特征向量，但是在'randomized'中，分解器会先生成多个随机向量，然后去检验这些随机向量中是否有任何一个符合分解需求，如果符合就保留这个随机变量，并基于这个随机向量来构建后续需要的向量空间。适合特征矩阵巨大，计算量庞大的情况。

##### 注意：random_state在其取值为'arpack'和'randomized'的时候起作用

### 使用PCA实现噪声过滤

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
digits=load_digits()
digits.data #数据
digits.target #目标值
digits.data.shape
```

```python
# 查看图像
def plot_digits(data):
    fig,axes=plt.subplots(4,10
                         ,figsize=(10,4)
                         ,subplot_kw={"xticks":[],"yticks":[]}
                        )
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap="binary_r")
        
plot_digits(digits.data)
```

### 人为加入噪音

```python
rng=np.random.RandomState(42)
# 在指定数据集中，随机抽取服从正态分布的数据
# 两个参数，分别为指定的数据集和抽取出来的正态分布的方差

noisy=rng.normal(digits.data,2)
noisy.shape
```

```python
# 查看加入噪声后的数据
plot_digits(noisy)
```

### 降维去噪

```python
# 分三步
pca=PCA(0.5,svd_solver='full').fit(noisy) #实例化并进行拟合
x_dr=pca.transform(noisy) #得到新的矩阵
x_dr.shape

# 一步到位
x_dr1=PCA(0.5,svd_solver='full').fit_transform(noisy)
x_dr1.shape
```

```python
without_noise=pca.inverse_transform(x_dr)
without_noise.shape
```

```python
plot_digits(without_noise)
```
